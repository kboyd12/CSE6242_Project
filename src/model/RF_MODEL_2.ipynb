```python
import pandas as pd
import glob
import numpy as np
import matplotlib.pyplot as plt
import pickle
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_curve, auc
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn import datasets
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
path = "../../data/raw/btsdelay/"
file_list = glob.glob(f"{path}*.parquet")

df = pd.concat([pd.read_parquet(i) for i in file_list])
```

```python
pd.set_option('display.max_columns', 200)
df.head()
```

```python
df['Season'] = np.where(df['Month'].isin([12,1,2]), 'Winter'
                        , np.where(df['Month'].isin([3,4,5]),'Spring'
                           , np.where(df['Month'].isin([6,7,8]),'Summer','Fall')))
```

Data is from Jan 2018 to July 2022

Option 1: One random forest model with origin, month, and airline as predictors

```python
#Data
reg_data = df[['Origin', 'Month','Airline','DepDelayMinutes','Year']].copy()
reg_data = reg_data[(reg_data['DepDelayMinutes']<1000)]
reg_data['Delay_Ind'] = np.where(reg_data['DepDelayMinutes']> 5,1,0)
reg_data_new = reg_data.sample(n=1000000, random_state=1234)
feature_cols = ['Origin', 'Month','Airline']

categorical_cols = ['Origin', 'Airline']
numerical_cols = ['Month']
feature_cols = categorical_cols + numerical_cols
```

```python
categorical_transformer = Pipeline(steps=[('onehot', OneHotEncoder(handle_unknown='ignore'))])
preprocessor = ColumnTransformer(transformers=[('cat', categorical_transformer, categorical_cols)],
                                 remainder='passthrough')

rf = Pipeline(steps=[('preprocessor', preprocessor),
                      ('classifier', RandomForestClassifier(max_depth=10,n_estimators=10,random_state=1234))])

X = reg_data_new[feature_cols]
y = reg_data_new['Delay_Ind']

model = rf.fit(X, y)
```

```python
param_grid = {
    'preprocessor__cat__onehot__handle_unknown': ['error', 'ignore'],
    'classifier__max_depth': [5, 10, 15,20],
    'classifier__min_samples_leaf': [5,10,20],
    'classifier__n_estimators': [10,25,30]
}

from sklearn.model_selection import GridSearchCV

grid_search = GridSearchCV(estimator=rf,
                          # param_grid=params,
                           param_grid=param_grid,
                           cv = 4,
                           n_jobs=-1, verbose=1, scoring="accuracy")
grid_search.fit(X, y)
grid_search.best_score_
```

```python
rf_best = grid_search.best_estimator_
rf_best
```

Updated Model with Hyperparameter Tuning

```python
rf_final = Pipeline(steps=[('preprocessor', preprocessor),
                      ('classifier', RandomForestClassifier(max_depth=20, min_samples_leaf=5 ,n_estimators=10,random_state=1234))])

model_final = rf_final.fit(X, y)
```

```python
Test = reg_data[~reg_data.index.isin(reg_data_new.index)].sample(n=10000, random_state=1234)
# Compute the false positive rate, true positive rate, and threshold
predictions_test = model_final.predict_proba(Test[feature_cols])
fpr, tpr, thresholds = roc_curve(Test['Delay_Ind'], predictions_test[:,1])

# Compute the area under the curve (AUC) score
roc_auc = auc(fpr, tpr)

# Plot the ROC curve
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic Example')
plt.legend(loc="lower right")
plt.show()
```

Option 2: Training multiple RF Forest Models from each Origin with Month, Airline, Dest as predictor (only 92 origins based on sample data)

```python
reg_data = df[['Origin', 'Month','Airline','DepDelayMinutes','Year','Dest']].copy()
reg_data = reg_data[(reg_data['DepDelayMinutes']<1000)]
reg_data['Delay_Ind'] = np.where(reg_data['DepDelayMinutes']> 5,1,0)

origin_airports = pd.DataFrame(reg_data['Origin'].value_counts()).reset_index()
origin_airports = origin_airports[origin_airports['Origin']>100000]
origin_airports['index']

models = []
model_Origin = []
for i in origin_airports['index']:
    air_data = []
    air_data = reg_data[reg_data['Origin']==i]
    air_data_new = air_data.sample(n=100000, random_state=1234)
    feature_cols = ['Month','Airline','Dest']

    categorical_cols = ['Airline','Dest']
    numerical_cols = ['Month']
    feature_cols = categorical_cols + numerical_cols
    
    categorical_transformer = Pipeline(steps=[('onehot', OneHotEncoder(handle_unknown='ignore'))])
    preprocessor = ColumnTransformer(transformers=[('cat', categorical_transformer, categorical_cols)],
                                 remainder='passthrough')

    rf = Pipeline(steps=[('preprocessor', preprocessor),
                      ('classifier', RandomForestClassifier(max_depth=20, min_samples_leaf=5 ,n_estimators=10,random_state=1234))])
    
    X = air_data_new[feature_cols]
    y = air_data_new['Delay_Ind']

    model = rf.fit(X, y)
    models += [model]
    model_Origin += [i]
```

```python
model_dict = {}
for i, model in enumerate(model_Origin):
    model_dict[model] = models[i]
```

```python
with open("models.pkl", "wb") as f:
    pickle.dump(model_dict, f)
```
